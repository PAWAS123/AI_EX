```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python

shop_data=pd.read_csv('shops.csv')
#shop_data
```

```python
names=shop_data.shop_name.to_list()

refined=[]
for name in names:
    name=name.lower()
    s1="".join(c for c in name if c.isalpha())
    refined.append(s1)
refined
ids=[(name,refined.index(name)) for name in refined if refined.count(name)>1]
ids
```

```python
import re
cities=[]
internet_shop=['Internet','Online']
for name in names:
    s=re.sub('[^a-zA-Z0-9\n\.]', ' ',name)
    l1=s.split()
    #print(l1)
    if l1[0]=='St.':
        city='St. Petersburg'
        cities.append(city)
    else:
        cities.append(l1[0])
        
cities_new=[city if 'Rostov' not in city else 'Rostov-Na-Donu' for city in cities ]
#cities_new
len(cities_new)
```

```python

locations=[]
geolocator = Nominatim(user_agent="pawas_app")
for city in cities_new:
    if 'Internet'==city or 'Digital'==city:
        locations.append('Online store')
    else:
        locations.append(geolocator.geocode(city))
        
#print(location.address)
```

```python
#locations
```

```python
binary_encode=[]
for item in cities_new:
    if 'Internet'==item or 'Digital'==item:
        binary_encode.append(1)
    else:
        binary_encode.append(0)

#binary_encode   
#len(binary_encode)
```

```python
shop_data['Locations']=locations
shop_data['Online shop']=binary_encode
shop_data
```
